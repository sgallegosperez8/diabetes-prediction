## Data-Preprocessing

Import diabetes data set 

```python
import pandas as pd
```

```python
diabetes = pd.read_csv("diabetes_prediction_dataset.csv")
```

## Data Wrangling
Checked the data for na and null values. Also filtered data sets by women and by age groups 18 and under, 19-30, 31-45, and 45 and above 

```python
diabetes_women = diabetes.query('gender == "Female"')
diabetes = diabetes_women.sort_values("age",ascending=True)
```

There is no na values

```python
diabetes_women.isna().sum()
```

There is no null values

```python
diabetes_women.isnull().sum()
```

Filtering by age groups 18 and under


```python
diabetes_w_18 = diabetes_women.query('age <= 18.0')
diabetes_w_18.sort_values("age",ascending=True)
#diabetes_w_18["age"].unique()

```

Create dataset with age group 19-30


```python
diabetes_w_19_30 = diabetes_women.query('age <= 20 and age >=19')
diabetes_w_19_30.sort_values("age",ascending=True)
```

Filter to age group 30 -45

```python
diabetes_w_31_45 = diabetes_women.query('age <= 45 and age >=31')
diabetes_w_31_45.sort_values("age",ascending=True)
```

Fitler to age group 46 and older 

```python
diabetes_w_46= diabetes_women.query('age >=46')
diabetes_w_46.sort_values("age",ascending=True)
```
